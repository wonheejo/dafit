```python
# Python, Numpy, OS libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
sns.set()
%matplotlib inline

import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
pio.templates.default = 'plotly_dark'
from plotly.subplots import make_subplots
```

```python
# Information about the columns and diabetes in general to understand about diabetes and the tests done to check for type1 and type2 diabetes
# Type 1 diabetes is often diagnosed more rapidly because it usually develops rapidly and involves high blood sugar levels and symptoms
# Type 2 diabetes develops often in over the course of years. Most common diabetes and usually develops during adulthood
# Diabetes can cause pregnancy complications and is usually recommended to test withtin 24~28 weeks of pregnancy.


# preg = Number of times pregnant
# plas = Plasma glucose concentration in a 2 hours in an oral glucose test
    # This measures how much the body is able to absorb the gluose(sugar). This is usually done to test diabetes during pregnency
# pres = Diastolic blood pressure(mm Hg) - this cannot be 0 because no one has 0 blood pressure when they are alive. 
# skin = Triceps skin folding thickness(mm)
# insu = 2 Hour serum insulin(mu U/ml) // Test is used to find out the levels of insulin in blood, Basically whether the patient is making enough insulin or not 
# mass = body mass index(weight in kgs) - BMI, it is considered overweight when BMI is greater than 25
# pedi = Diabetes pedigree function
# age = Age of patient
# class = Target(whether patient has diabetes or not)
```

```python
train = pd.read_csv('dataset_37_diabetes.csv')
```

```python
zeros = (train==0).astype(int).sum(axis=0)
zeros
```

```python
# Change the 0 values to NaNs to fill in the missing values
cols = ['mass', 'pres', 'plas']
train[cols] = train[cols].replace({0: np.nan})
```

```python
# To fill the various needed missing values with the average of the 'age' and 'mass'. This is solely based on my reference and research. 
train['mass'].fillna(train.groupby('age')['mass'].transform('mean'), inplace=True)
train['pres'].fillna(train.groupby(['mass'])['pres'].transform('mean'), inplace=True)
train['pres'].fillna(train.groupby(['age'])['pres'].transform('mean'), inplace=True)
train['plas'].fillna(train.groupby(['age'])['plas'].transform('mean'), inplace=True)
```

```python
train.isna().sum()
```

```python
# Using matplotlib to look at the numbers between the tested positive and tested negative
plt.subplots(figsize=(15, 10))
sns.countplot(x='class', data=train)
```

```python
# Using matplotlibs histogram function to look at all of the 8 columns about their counts
columns = train.columns[:8]
plt.subplots(figsize=(18, 15))
length = len(columns)
for i, j in itertools.zip_longest(columns, range(length)):
    plt.subplot((length/2), 3, j+1)
    plt.subplots_adjust(wspace=0.2, hspace=0.5)
    train[i].hist(bins=20, edgecolor='black')
    plt.title(i)
plt.show()
```

```python
# Looking at matplotlibs histogram function to look at the values of those tested positive of diabetes
train1 = train[train['class'] == 'tested_positive']
columns = train.columns[:8]
plt.subplots(figsize=(18, 15))
length = len(columns)
for i, j in itertools.zip_longest(columns, range(length)):
    plt.subplot((length/2), 3, j+1)
    plt.subplots_adjust(wspace=0.2, hspace=0.5)
    train1[i].hist(bins=20, edgecolor='black')
    plt.title(i)
plt.show()
```

```python
# Using plotly to look at each of the columns seperately compared to each other. 
fig = go.Figure()
fig.add_trace(go.Histogram(histfunc='count', x=train['class']))
fig.update_layout(title='Diabetes outcome')
fig.show()
```

```python
# Another way of using plotly to show the same bar graph
fig = px.histogram(train, x='class')
fig.show()
```

```python
# Using plotly to show all the columns in one output using subplot function
# The reason i used plotly was because of its extensive customization available to the user. 
columns = train.columns
length = len(columns)
fig = make_subplots(rows=3, cols=3, subplot_titles=('Pregnancy', 'Plasma', 'Pressure', 'Skin', 'Insulin', 'BMI', 'Pedi', 'Age', 'Class'))
for i, j in itertools.zip_longest(columns, range(length)):

    fig.add_trace(
        go.Histogram(histfunc='count', x=train[i], showlegend=False),
        row=j//3+1, 
        col=(j%3)+1)
    
fig.update_layout(title='Plots of all columns', height=800, width=950, bargap=0.1)
fig.show()
```

```python
# Using plotly to show the graphs of values of patients who are positive to diabetes
train1 = train[train['class'] == 'tested_positive']
columns=train1.columns[:8]
length = len(columns)
fig = make_subplots(rows=3, cols=3, subplot_titles=('Pregnancy', 'Plasma', 'Pressure', 'Skin', 'Insulin', 'BMI', 'Pedi', 'Age'))
for i, j in itertools.zip_longest(columns, range(length)):
    fig.add_trace(go.Histogram(x=train1[i], showlegend=False),
                 row=j//3+1, col=j%3+1)
fig.update_layout(height=800, width=950, title='Multiple plots of diabetes patients', bargap=0.1)
fig.show()
```
