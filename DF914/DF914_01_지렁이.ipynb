```python
import pandas as pd
import numpy as np
from datetime import datetime
pd.set_option('display.max_rows', 500)
```

```python
# Dropped the following missing columns due to the fact that they wont be much of a help for the analysis and that they are practically nearly all missing!
train = train.drop(['Phone', 'Real Email', 'Notes', 'Opt Out', 'OriginalState'], axis=1)
```

```python
# Used to find the specific row in which it has a null value. In this code I have searched for the specific row in 'ClientId' Column and found out that this row
# also has the missing LastPurchaseDate and FirstPurchaseDate so I will just remove this row as it has missing State, Country and City. 
null_columns = train.columns[train.isnull().any()]
train[train['ClientId'].isnull()][null_columns]
```

```python
# Removed the single row which doesnt have client id
train = train.dropna(how='any', subset=['ClientId', 'City'])
```

```python
# Parsing the date format from its original format for easier visualization. 
train['Lastdate'] = train['LastPurchaseDate'].apply(lambda x: x.split())
train['Firstdate'] = train['FirstPurchaseDate'].apply(lambda x: x.split())
```

```python
# Created two new columns after splitting them into just Year month date format
train['FirstYmd'] = train['Firstdate'].apply(lambda x: x[0])
train['LastYmd'] = train['Lastdate'].apply(lambda x: x[0])
```

```python
train.head()
```

```python
# Created 2 new columns for hour and minute using split function
train['Firsthm'] = train['Firstdate'].apply(lambda x: x[1])
train['Lasthm'] = train['Lastdate'].apply(lambda x: x[1])
```

```python
train.head()
```

```python
# Dropped the original LastPurchaseDate and FirstPurchaseDate columns as well as Lastdate and Firstdate now that i have seperate columns for Ymd and Hm
train = train.drop(['LastPurchaseDate', 'FirstPurchaseDate', 'Lastdate', 'Firstdate'], axis=1)
```

```python
train.head()
```

```python
# Converted the FirstYmd and LastYmd columns into datetime so I can calculate the difference of days between them
train['FirstYmd'] = pd.to_datetime(train['FirstYmd'])
train['LastYmd'] = pd.to_datetime(train['LastYmd'])
```

```python
# Calculating the difference of days between them
train['totaldays'] = train['LastYmd'] - train['FirstYmd']
train['totaldays']
```

```python
# Filling in the missing values for the 83 buyers. I decided to use the median value of each states because after looking at the values or buyers who spent less than
# a day, they all seem to have bought the same thing as if there was a sale on a particular item.

train['# of Orders Placed'].fillna(first['# of Orders Placed'], inplace=True)
train['# Different products (SKUs) purchased'].fillna(second['# Different products (SKUs) purchased'], inplace=True)
train['# of Items Purchased'].fillna(third['# of Items Purchased'], inplace=True)
train['Max Orders for one SKU'].fillna(fourth['Max Orders for one SKU'], inplace=True)
```

```python
# Used to find the median value of each of the columns and changed their format into integer from float because the .transform('median') does not seem to work
# due to the fact that it is not considered 'numeric'
first['# of Orders Placed'] = train.groupby('State')['# of Orders Placed'].median().astype(int)
second['# Different products (SKUs) purchased'] = train.groupby('State')['# Different products (SKUs) purchased'].median()
third['# of Items Purchased'] = train.groupby('State')['# of Items Purchased'].median().astype(int)
fourth['Max Orders for one SKU'] = train.groupby('State')['Max Orders for one SKU'].median().astype(int)
```

```python
# Converted the column into float type from string object type
train['Total Value of all Orders Placed'] = train['Total Value of all Orders Placed'].astype(float)
```

```python
# Once converted to float i was able to use transform function to replace all the null value to median value
train['Total Value of all Orders Placed'].fillna(train.groupby('State')['Total Value of all Orders Placed'].transform('median'), inplace=True)
```
